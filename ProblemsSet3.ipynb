```python
QUESTION 6: Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("us-marriages-divorces-1867-2014.csv")

# Filter the data for the years between 1867 and 2014
data = data[(data['Year'] >= 1867) & (data['Year'] <= 2014)]

# Create a line plot
data.plot(x='Year', y=['Marriages_per_1000', 'Divorces_per_1000'], figsize=(10, 6))

# Labeling the plot
plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.xlabel('Year')
plt.ylabel('Number per Capita')
plt.legend(['Marriages per 1000', 'Divorces per 1000'])

plt.show()
```

```python
QUESTION 7:Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("us-marriages-divorces-1867-2014.csv")

# Filter the data
years = [1900, 1950, 2000]
data_selected_years = data[data['Year'].isin(years)]

# Calculate marriages and divorces per capita for the selected years using lambda functions
data_selected_years['Marriages_per_capita'] = data_selected_years.apply(lambda row: row['Marriages'] / row['Population'], axis=1)
data_selected_years['Divorces_per_capita'] = data_selected_years.apply(lambda row: row['Divorces'] / row['Population'], axis=1)

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(years, data_selected_years['Marriages_per_capita'], width=0.4, align='center', label='Marriages per Capita', color='blue')
plt.bar([year + 0.4 for year in years], data_selected_years['Divorces_per_capita'], width=0.4, align='center', label='Divorces per Capita', color='red')

# Labeling the plot
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
plt.xlabel('Year')
plt.ylabel('Number per Capita')
plt.xticks(years)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
QUESTION 8: Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("actor_kill_counts.csv")

# Sort the data by kill count in descending order
data_sorted = data.sort_values(by='Count', ascending=False)

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(data_sorted['Actor'], data_sorted['Count'], color='red')

# Labeling the plot
plt.title('Deadliest Actors in Hollywood')
plt.xlabel('Count')
plt.ylabel('Actor')

# Label each bar with the corresponding actor's name
for i, v in enumerate(data_sorted['Count']):
    plt.text(v + 1, i, str(v), color='black', va='center')

plt.tight_layout()
plt.show()
```

```python
QUESTION 9: Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("roman-emperor-reigns.csv")

#total number of emperors
total_emperors = len(data)

# Count of emperors who were assassinated
assassinated_emperors = data[data['Cause_of_Death'] == 'Assassinated'].shape[0]

# Calculate the fraction of emperors who were assassinated
fraction_assassinated = assassinated_emperors / total_emperors

# Create a DataFrame for the pie chart
categories = ["Assassinated", "Died in captivity", "Executed", "Illness", "Killed in battle", "Natural causes", "Possibly assassinated", "Suicide"]
assassination_counts = [data[data['Cause_of_Death'] == category].shape[0] for category in categories]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(assassination_counts, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title('Fraction of Roman Emperors Assassinated')

plt.axis('equal')

plt.show()
```

```python
QUESTION 10: Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("arcade-revenue-vs-cs-doctorates.csv")

# Filter data
filtered_data = data[(data['Year'] >= 2000) & (data['Year'] <= 2009)]

# Create a scatter plot
plt.figure(figsize=(10, 6))
for year, group in filtered_data.groupby('Year'):
    plt.scatter(group['Total Arcade Revenue (billions)'], group['Computer Science Doctorates Awarded (US)'], label=year)

# Labeling
plt.title('Relationship between Arcade Revenue and Computer Science PhDs (2000-2009)')
plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Computer Science Doctorates Awarded (US)')
plt.legend(title='Year')

plt.grid(True)
plt.tight_layout()
plt.show()
```
